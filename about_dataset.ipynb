```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import matplotlib as mpl
import cv2
```

```python
count_train_ayrshire = len(os.listdir("/content/drive/MyDrive/Cattle Classification/data/train/Ayrshire cattle"))
count_train_brown_swiss = len(os.listdir("/content/drive/MyDrive/Cattle Classification/data/train/Brown Swiss cattle"))
count_train_holstein = len(os.listdir("/content/drive/MyDrive/Cattle Classification/data/train/Holstein Friesian cattle"))
count_train_jersey = len(os.listdir("/content/drive/MyDrive/Cattle Classification/data/train/Jersey cattle"))
count_train_red_dane = len(os.listdir("/content/drive/MyDrive/Cattle Classification/data/train/Red Dane cattle"))

count_val_ayrshire = len(os.listdir("/content/drive/MyDrive/Cattle Classification/data/val/Ayrshire cattle"))
count_val_brown_swiss = len(os.listdir("/content/drive/MyDrive/Cattle Classification/data/val/Brown Swiss cattle"))
count_val_holstein = len(os.listdir("/content/drive/MyDrive/Cattle Classification/data/val/Holstein Friesian cattle"))
count_val_jersey = len(os.listdir("/content/drive/MyDrive/Cattle Classification/data/val/Jersey cattle"))
count_val_red_dane = len(os.listdir("/content/drive/MyDrive/Cattle Classification/data/val/Red Dane cattle"))

count_test_ayrshire = len(os.listdir("/content/drive/MyDrive/Cattle Classification/data/test/Ayrshire cattle"))
count_test_brown_swiss = len(os.listdir("/content/drive/MyDrive/Cattle Classification/data/test/Brown Swiss cattle"))
count_test_holstein = len(os.listdir("/content/drive/MyDrive/Cattle Classification/data/test/Holstein Friesian cattle"))
count_test_jersey = len(os.listdir("/content/drive/MyDrive/Cattle Classification/data/test/Jersey cattle"))
count_test_red_dane = len(os.listdir("/content/drive/MyDrive/Cattle Classification/data/test/Red Dane cattle"))

total_ayrshire = count_train_ayrshire + count_val_ayrshire + count_test_ayrshire
total_browns_swiss = count_train_brown_swiss + count_val_brown_swiss + count_test_brown_swiss
total_holstein = count_train_holstein + count_val_holstein + count_test_holstein
total_jersey = count_train_jersey + count_val_jersey + count_test_jersey
total_red_dane = count_train_red_dane + count_val_red_dane + count_test_red_dane
```

```python
breeds = ('Ayrshire', 'Brown Swiss', 'Holstein', "Jersey", "Red Dane")
data = {
    'Train': np.array([count_train_ayrshire, count_train_brown_swiss, count_train_holstein, count_train_jersey, count_train_red_dane]),
    'Validation': np.array([count_val_ayrshire, count_val_brown_swiss, count_val_holstein, count_val_jersey, count_val_red_dane]),
    'Test': np.array([count_test_ayrshire, count_test_brown_swiss, count_test_holstein, count_test_jersey, count_test_red_dane])
}
width = 0.6
fig, ax = plt.subplots()
bottom = np.zeros(5)

for split, values in data.items():
    p = ax.bar(breeds, values, width, bottom=bottom, label=split)
    bottom += values

    ax.bar_label(p, label_type='center')

ax.set_title('Dataset Distribution')
ax.legend(loc='upper right', title="Split")

plt.show()
```

```python

image_paths = ["/content/drive/MyDrive/Cattle Classification/data/train/Ayrshire cattle/Ayrshirecattle0.jpg", "/content/drive/MyDrive/Cattle Classification/data/train/Ayrshire cattle/Ayrshirecattle100.jpg",
               "/content/drive/MyDrive/Cattle Classification/data/train/Brown Swiss cattle/BrownSwisscattle100.jpg","/content/drive/MyDrive/Cattle Classification/data/train/Brown Swiss cattle/BrownSwisscattle104_c.jpg",
               "/content/drive/MyDrive/Cattle Classification/data/train/Brown Swiss cattle/BrownSwisscattle104_c.jpg", "/content/drive/MyDrive/Cattle Classification/data/train/Holstein Friesian cattle/HolsteinFriesiancattle103_c.jpg",
               "/content/drive/MyDrive/Cattle Classification/data/train/Jersey cattle/Jerseycattle100.jpg", "/content/drive/MyDrive/Cattle Classification/data/train/Jersey cattle/Jerseycattle105_c.jpg",
               "/content/drive/MyDrive/Cattle Classification/data/train/Red Dane cattle/RedDanecattle100.jpg",  "/content/drive/MyDrive/Cattle Classification/data/train/Red Dane cattle/RedDanecattle107.jpg"]

fig, axes = plt.subplots(2, 5, figsize=(15, 6))
axes = axes.ravel()

for i, path in enumerate(image_paths):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    axes[i].imshow(img)

    axes[i].set_title(os.path.basename(path), fontsize=8)
    axes[i].axis("off")

plt.tight_layout()
plt.show()
```
